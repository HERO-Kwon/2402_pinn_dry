```python
matrix = mat_file['F']
# 히트맵 그리기 함수
def plot_heatmap(matrix):
    plt.figure(figsize=(8, 6))
    plt.imshow(matrix, cmap='viridis', interpolation='nearest')
    plt.colorbar()  # 색상 막대 추가
    plt.title('Heatmap of 2D Matrix')
    plt.xlabel('Column')
    plt.ylabel('Row')
    plt.show()

# 히트맵 그리기
plot_heatmap(matrix)
```
